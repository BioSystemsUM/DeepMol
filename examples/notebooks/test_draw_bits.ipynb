```python
from rdkit.Chem import MolFromSmiles

mol = MolFromSmiles("CC(C)(C)C(O)C(O)=O")
```

```python
from deepmol.compound_featurization import MorganFingerprint

morgan_fingerprint = MorganFingerprint()
morgan_fingerprint.draw_bits(mol, [1, 114, 227])
```

```python
morgan_fingerprint.draw_bit(mol, 1)
```

```python
from deepmol.compound_featurization import MACCSkeysFingerprint

maccs_keys_fingerprints = MACCSkeysFingerprint()
maccs_keys_fingerprints.draw_bit(mol, 164)
```

```python
from deepmol.compound_featurization import RDKFingerprint

rdk_fingerprints = RDKFingerprint()
rdk_fingerprints.draw_bit(mol, 56)
```

```python
rdk_fingerprints.draw_bits(mol, [56, 61, 137])
```
